```python
import pandas as pd
import random
```

```python
details_df = pd.read_csv("Details.csv")
placement_df = pd.read_csv("Placement.csv")

print(details_df,placement_df)
```

```python
all = pd.merge(details_df,placement_df,on="Roll No.",how="outer")

print(all)
```

```python
#only places students data

placed = all[all["Placed"]=="Yes"].copy()

print(placed)
```

```python
#reset index

placed.reset_index(drop=True, inplace=True)

print(placed)
```

```python
#assigning salary based on CPI

salary1 = []


for i in range(28):
    cpi = placed.loc[i].at["CPI"]

    if(cpi>=8):
        salary = random.randint(65000,70000)
        salary1.append(salary)
        
    elif(cpi>=7):
        salary = random.randint(55000,60000)
        salary1.append(salary)
        
    elif(cpi>=6):
        salary = random.randint(50000,55000)
        salary1.append(salary)
        
    else:
        salary = random.randint(45000,50000)
        salary1.append(salary)
        
        
```

```python
placed.insert(17,"Salary",salary1,True)

#remove specific column 
#placed.drop("Salary",axis="columns",inplace=True)

print(placed)
```

```python
#data visualization

import matplotlib.pyplot as plt
```

```python
x = placed['CPI']
y = placed['Salary']
z = placed['Roll No.']
```

```python
#Scatter plot

plt.xlabel("CPI")
plt.ylabel("Salary")
plt.title("CPI VS SALARY")
plt.plot(placed["CPI"],marker='o')
plt.show()
```

```python
#rollno vs cpi
plt.xlabel("Roll No")
plt.ylabel("CPI")
plt.title("Roll No VS CPI")
plt.plot(z,x,marker='o')
plt.show()
```

```python
plt.plot(x, marker='o')
plt.plot(y,marker='o')
```

```python
#-----normalization-----------

#min-max normalaization

normalised_s = []
for i in range(28):
    s = placed.loc[i].at["Salary"]
    s = s/10000
    normalised_s.append(s)
```

```python
plt.xlabel("Index")
plt.ylabel("CPI/Salary")
plt.title("Normalised Salary VS CPI")
plt.plot(x,marker="o")
plt.plot(normalised_s,marker='o')
plt.show()
```

```python
plt.hist(placed["CPI"])
plt.hist(placed["Salary"])
```

```python
#pie chart

plt.pie(y)
```

```python
#boxPlot

plt.boxplot(placed["Salary"])
```

```python
#bar chart

plt.bar(x,y)
```

```python
#scatter graph

plt.scatter(placed["CPI"],placed["Salary"],c=placed["HSC Result"])
plt.colorbar()

plt.xlabel("CPI")
plt.ylabel("Salary")
```
