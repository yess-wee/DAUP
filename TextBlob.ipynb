```python
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
text_data = """ rin teh quinte vilege of Willowbrook, nestled between rolling hills and vast fields, the local bakery stood as a cornerstone of community life. Each morning, the scent of freshly baked bread and pastries filled the air, drawing people from all walks of life into its warm, welcoming embrace. Here, amidst the clatter of teacups and the murmur of friendly conversation, differences melted away, friendships were forged, and the simple joys of life were celebrated. It was in this little bakery that the true heart of Willowbrook beat strongest, a testament to the power of good food and good company. """
```

```python
#tokanization

blob = TextBlob(text_data)
tokens = blob.words
```

```python
print("Tokens : ",tokens)
```

```python
sentiment = blob.sentiment
print("sentiment",sentiment)
```

```python
tags = blob.tags
print("tags",tags)
```

```python
wordcloud = WordCloud().generate(text_data)
```

```python
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
